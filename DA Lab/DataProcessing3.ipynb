```python
import pandas as pd
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
import numpy as np

df = pd.read_csv('D:/210968186/B1/movielens/ratings.csv')
df
```

```python
df2 = pd.pivot_table(df,index="userId",columns="movieId",values="rating",fill_value=0)
df2
```

```python
array = pd.DataFrame(data=pairwise_distances(df2, Y=None, metric='cosine')[24]).nsmallest(6,columns=0).index
print("Most similar users are:")
for values in array:
    if(values != 24):
        print('User',values+1)
```
