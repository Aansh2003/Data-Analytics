```python
list_4 = []
list_6 = []
list_8 = []
for value in range(0,ran):
    if df['cyl'][value] == 4:
        list_4.append(df['mpg'][value])
    elif df['cyl'][value] == 6:
        list_6.append(df['mpg'][value])
    elif df['cyl'][value]==8:
        list_8.append(df['mpg'][value])
```
